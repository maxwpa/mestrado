```r
rede <- read_excel("painel_skills_rede.xlsx")
oes <- read_excel("oesm-v2.xlsx")
```

```r
names(oes) <- gsub("\\*", "", names(oes))

oes <- oes %>%
  rename(
    O_NET_SOC_Code = `ONET-SOC Code`,
    Version_Year = YEAR
  )
```

```r
oes <- oes %>%
  mutate(

    A_MEAN = ifelse(
      is.na(A_MEAN),
      H_MEAN * 2080,
      A_MEAN
    )

  )
```

```r
painel <- rede %>%
  left_join(
    oes,
    by = c(
      "O_NET_SOC_Code",
      "Version_Year"
    )
  )
```

## REGRESSÃO EM DADOS EM PAINEL

### Interpolação dentro da ocupação

```r
vars_interp <- c(

  "Comunidade_1_Comunicacao",
  "Comunidade_2_Analitica",
  "Comunidade_3_Gestao",
  "Comunidade_4_Social",
  "Comunidade_5_Tecnica",
  "Comunidade_6_Recursos",

  "Systems_Evaluation_Level",
  "Reading_Comprehension_Level",
  "Complex_Problem_Solving_Level",
  "Systems_Analysis_Level",
  "Active_Learning_Level",
  "Troubleshooting_Level",
  "Repairing_Level",
  "Operations_Monitoring_Level"

)

painel <- painel %>%
  arrange(
    O_NET_SOC_Code,
    Version_Year
  ) %>%
  group_by(O_NET_SOC_Code) %>%
  mutate(
    across(
      all_of(vars_interp),
      ~ zoo::na.approx(
        .,
        na.rm = FALSE
      )
    )
  ) %>%
  ungroup()
```

### Variáveis dependentes

```r
painel <- painel %>%
  mutate(

    log_emprego =
      log(TOT_EMP + 1),

    log_salario =
      log(A_MEAN)

  )
```

### Painel

```r
painel_plm <- pdata.frame(
  painel,
  index = c(
    "O_NET_SOC_Code",
    "Version_Year"
  )
)
```

### MODELO 1 — Comunidades → Salário

```r
m_salario_com <- plm(

  log_salario ~

    Comunidade_1_Comunicacao +
    Comunidade_2_Analitica +
    Comunidade_3_Gestao +
    Comunidade_4_Social +
    Comunidade_5_Tecnica +
    Comunidade_6_Recursos,

  data = painel_plm,

  model = "within"
)

summary(m_salario_com)
```

### MODELO 2 — Habilidades centrais → Salário

```r
m_salario_hubs <- plm(

  log_salario ~

    Systems_Evaluation_Level +
    Reading_Comprehension_Level +
    Complex_Problem_Solving_Level +
    Systems_Analysis_Level +
    Active_Learning_Level +
    Troubleshooting_Level +
    Repairing_Level +
    Operations_Monitoring_Level,

  data = painel_plm,

  model = "within"
)

summary(m_salario_hubs)
```

### MODELO 3 — Comunidades → Emprego

```r
m_emprego_com <- plm(

  log_emprego ~

    Comunidade_1_Comunicacao +
    Comunidade_2_Analitica +
    Comunidade_3_Gestao +
    Comunidade_4_Social +
    Comunidade_5_Tecnica +
    Comunidade_6_Recursos,

  data = painel_plm,

  model = "within"
)

summary(m_emprego_com)
```

### MODELO 4 — Habilidades centrais → Emprego

```r
m_emprego_hubs <- plm(

  log_emprego ~

    Systems_Evaluation_Level +
    Reading_Comprehension_Level +
    Complex_Problem_Solving_Level +
    Systems_Analysis_Level +
    Active_Learning_Level +
    Troubleshooting_Level +
    Repairing_Level +
    Operations_Monitoring_Level,

  data = painel_plm,

  model = "within"
)

summary(m_emprego_hubs)
```

### Erros robustos

```r
print('SALÁRIOS')
coeftest(
  m_salario_com,
  vcov = vcovHC(
    m_salario_com,
    type = "HC1"
  )
)
```

```r
print('SALÁRIOS')
coeftest(
  m_salario_hubs,
  vcov = vcovHC(
    m_salario_hubs,
    type = "HC1"
  )
)
```

```r
print('EMPREGOS')
coeftest(
  m_emprego_com,
  vcov = vcovHC(
    m_emprego_com,
    type = "HC1"
  )
)
```

```r
print('EMPREGO')
coeftest(
  m_emprego_hubs,
  vcov = vcovHC(
    m_emprego_hubs,
    type = "HC1"
  )
)
```

### Hausman

```r
m_re <- plm(

  log_salario ~

    Systems_Evaluation_Level +
    Reading_Comprehension_Level +
    Complex_Problem_Solving_Level +
    Systems_Analysis_Level,

  data = painel_plm,

  model = "random"
)

phtest(
  m_salario_hubs,
  m_re
)
```
