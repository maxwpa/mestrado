```r
library(tidyverse)

library(psych)
library(GPArotation)

library(nFactors)

library(corrplot)

library(reshape2)

library(lavaan)
library(semTools)

library(blavaan)
library(bayesplot)
```

```r
efa_data_raw <- read.delim(
  "adequacy-c.txt",
  check.names = FALSE
)

names(efa_data_raw) <- gsub(
  "\\.+",
  "_",
  names(efa_data_raw)
)

id_cols <- c(
  "O_NET_SOC_Code",
  "Title",
  "Version_Year"
)

skill_vars <- setdiff(
  names(efa_data_raw),
  id_cols
)

efa_global <- efa_data_raw %>%
  select(all_of(skill_vars))
```

### 1. Solução com 6 Fatores

```r
efa6 <- fa(
  efa_global,
  nfactors = 6,
  fm = "minres",
  rotate = "promax"
)

print(
  efa6,
  sort = TRUE,
  cut = .30
)
```

### 2. Extrair cargas

```r
loadings_6 <- as.data.frame(
  unclass(efa6$loadings)
)

round(loadings_6, 3)
```

### 3. Comunalidades, R e Q

```r
communalities_6 <- tibble(
  skill = rownames(loadings_6),
  h2 = efa6$communality
)

communalities_6 %>%
  arrange(h2)
```

### 4. EFA longitudinal por ano

```r
years <- sort(
  unique(efa_data_raw$Version_Year)
)

efa_by_year <- list()

for(y in years){

  temp <- efa_data_raw %>%
    filter(Version_Year == y) %>%
    select(all_of(skill_vars))

  efa_by_year[[as.character(y)]] <- fa(
    temp,
    nfactors = 6,
    fm = "minres",
    rotate = "promax"
  )

}
```

```r
efa_by_year
```

### 5. Congruência de Tucker

```r
factor.congruence(
  efa_by_year[["2017"]],
  efa_by_year[["2019"]]
)

factor.congruence(
  efa_by_year[["2019"]],
  efa_by_year[["2021"]]
)

factor.congruence(
  efa_by_year[["2021"]],
  efa_by_year[["2023"]]
)

factor.congruence(
  efa_by_year[["2023"]],
  efa_by_year[["2025"]]
)
```

```r
print(factor.congruence(efa_by_year))
```

### 6. Carga Fatorial por Ano

```r
loadings_year <- lapply(
  efa_by_year,
  function(x){
    as.data.frame(unclass(x$loadings))
  }
)

loadings_year
```

```r
cargas_long <- bind_rows(
  lapply(names(loadings_year), function(y){

    loadings_year[[y]] %>%
      tibble::rownames_to_column("skill") %>%
      pivot_longer(
        starts_with("MR"),
        names_to = "factor",
        values_to = "loading"
      ) %>%
      mutate(year = y)

  })
)

print(cargas_long, n = Inf, width = Inf)
```

### 7. Modelo SEM 6 Fatores

```r
model_esem6 <- '

efa("block1")*F1 +
efa("block1")*F2 +
efa("block1")*F3 +
efa("block1")*F4 +
efa("block1")*F5 +
efa("block1")*F6 =~

Science_Level +                        
Programming_Level +                      
Mathematical_Reasoning_Level +                   
Reading_Comprehension_Level +                  
Complex_Problem_Solving_Level +                
Operations_Analysis_Level +                      
Technology_Design_Level +          
Critical_Thinking_Level +                
Active_Learning_Level +                      
Writing_Level +                       
Systems_Analysis_Level +                    
Judgment_and_Decision_Making_Level +                 
Systems_Evaluation_Level +                
Speaking_Level +                 
Persuasion_Level +                
Negotiation_Level +                
Social_Perceptiveness_Level +                 
Service_Orientation_Level +                  
Coordination_Level +                 
Time_Management_Level +                
Management_of_Personnel_Resources_Level +    
Active_Listening_Level +                 
Monitoring_Level +               
Repairing_Level +          
Equipment_Maintenance_Level +            
Troubleshooting_Level +          
Equipment_Selection_Level +           
Operations_Monitoring_Level +           
Operation_and_Control_Level +          
Quality_Control_Analysis_Level +           
Installation_Level +           
Management_of_Financial_Resources_Level +   
Management_of_Material_Resources_Level +     
Instructing_Level +
Learning_Strategies_Level 

'
```

### 8. Ajuste Fatorial

```r
fit_esem6 <- cfa(
    model_esem6,
    data = efa_global,
    estimator = "MLR",
    std.lv = TRUE
)

fitMeasures(
  fit_esem6,
  c("chisq","df","pvalue","cfi","tli","rmsea","rmsea.ci.lower","rmsea.ci.upper","srmr")
)
```

### 9. Loadings

```r
loadings <- parameterEstimates(fit_esem6)

loadings %>%
  filter(op == "=~") %>%
  as.data.frame() %>%
  print()
```
