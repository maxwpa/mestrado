```python
import pandas as pd

onet = pd.read_csv(r"onet-v5.csv", sep=",", low_memory=False)
oesm = pd.read_excel("oesm.xlsx")
```

```python
versoes = [
    22.1,
    24.1,
    26.1,
    28.1,
    30.1
]

onet = onet[
    onet["Version"].isin(versoes)
].copy()
```

```python
len(onet)
```

```python
chaves_oesm = oesm[['O*NET-SOC Code', 'YEAR']].drop_duplicates()

chaves_oesm = chaves_oesm.rename(columns={'YEAR': 'Version Year'})

onet['Version Year'] = onet['Version Year'].astype(str)
chaves_oesm['Version Year'] = chaves_oesm['Version Year'].astype(str)

onet = onet.merge(
    chaves_oesm,
    on=['O*NET-SOC Code', 'Version Year'],
    how='inner'
)
```

```python
len(onet)
```

```python
elementos = onet[['Element ID', 'Element Name', 'Type']].drop_duplicates()

nomes_duplicados = elementos[
    elementos.duplicated(subset='Element Name', keep=False)
]

nomes_duplicados = nomes_duplicados.sort_values(
    ['Element Name', 'Element ID', 'Type']
).reset_index(drop=True)

print(nomes_duplicados)
```

```python
onet.loc[
    (onet['Element ID'] == '2.A.1.e') & 
    (onet['Element Name'] == 'Mathematics'),
    'Element Name'
] = 'Mathematical Reasoning'
```
