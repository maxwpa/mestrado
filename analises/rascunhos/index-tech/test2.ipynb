```r
knowledge <- read_excel("Knowledge.xlsx")
skills <- read_excel("Skills.xlsx")
activities <- read_excel("Work_Activities.xlsx")
ete <- read_excel("ETE.xlsx")
job <- read_excel("Job_Zones.xlsx")
context <- read_excel("Work_Context.xlsx")
```

```r
# Padronizar nomes
knowledge <- clean_names(knowledge)
ete <- clean_names(ete)
job <- clean_names(job)
```

```r
knowledge_im <- knowledge %>% filter(scale_id == "IM")
knowledge_lv <- knowledge %>% filter(scale_id == "LV")

skills_im <- skills %>% filter(scale_id == "IM")
skills_lv <- skills %>% filter(scale_id == "LV")

activities_im <- activities %>% filter(scale_id == "IM")
activities_lv <- activities %>% filter(scale_id == "LV")
```

```r
knowledge_im_wide <- knowledge_im %>%
  pivot_wider(names_from = element_name, values_from = data_value)

knowledge_lv_wide <- knowledge_lv %>%
  pivot_wider(names_from = element_name, values_from = data_value)

skills_im_wide <- skills_im %>%
  pivot_wider(names_from = element_name, values_from = data_value)

skills_lv_wide <- skills_lv %>%
  pivot_wider(names_from = element_name, values_from = data_value)

activities_im_wide <- activities_im %>%
  pivot_wider(names_from = element_name, values_from = data_value)

activities_lv_wide <- activities_lv %>%
  pivot_wider(names_from = element_name, values_from = data_value)
```

```r
ete_wide <- ete %>%
  pivot_wider(names_from = element_name, values_from = data_value)

context_wide <- context %>%
  pivot_wider(names_from = element_name, values_from = data_value)
```

```r
knowledge_im_wide <- knowledge_im %>%
  select(o_net_soc_code, element_name, data_value) %>%
  pivot_wider(names_from = element_name, values_from = data_value) %>%
  group_by(o_net_soc_code) %>%
  summarise(across(everything(), ~ mean(.x, na.rm = TRUE)))

skills_im_wide <- skills_im %>%
  select(o_net_soc_code, element_name, data_value) %>%
  pivot_wider(names_from = element_name, values_from = data_value) %>%
  group_by(o_net_soc_code) %>%
  summarise(across(everything(), ~ mean(.x, na.rm = TRUE)))

activities_im_wide <- activities_im %>%
  select(o_net_soc_code, element_name, data_value) %>%
  pivot_wider(names_from = element_name, values_from = data_value) %>%
  group_by(o_net_soc_code) %>%
  summarise(across(everything(), ~ mean(.x, na.rm = TRUE)))
```

```r
knowledge_lv_wide <- knowledge_lv %>%
  select(o_net_soc_code, element_name, data_value) %>%
  pivot_wider(names_from = element_name, values_from = data_value) %>%
  group_by(o_net_soc_code) %>%
  summarise(across(everything(), ~ mean(.x, na.rm = TRUE)))

skills_lv_wide <- skills_lv %>%
  select(o_net_soc_code, element_name, data_value) %>%
  pivot_wider(names_from = element_name, values_from = data_value) %>%
  group_by(o_net_soc_code) %>%
  summarise(across(everything(), ~ mean(.x, na.rm = TRUE)))

activities_lv_wide <- activities_lv %>%
  select(o_net_soc_code, element_name, data_value) %>%
  pivot_wider(names_from = element_name, values_from = data_value) %>%
  group_by(o_net_soc_code) %>%
  summarise(across(everything(), ~ mean(.x, na.rm = TRUE)))
```

```r
knowledge_combined <- knowledge_im_wide
knowledge_combined[,-1] <- knowledge_im_wide[,-1] * knowledge_lv_wide[,-1]

skills_combined <- skills_im_wide
skills_combined[,-1] <- skills_im_wide[,-1] * skills_lv_wide[,-1]

activities_combined <- activities_im_wide
activities_combined[,-1] <- activities_im_wide[,-1] * activities_lv_wide[,-1]
```

```r
ete_wide <- ete %>%
  select(o_net_soc_code, element_name, data_value) %>%
  pivot_wider(names_from = element_name, values_from = data_value) %>%
  group_by(o_net_soc_code) %>%
  summarise(across(everything(), ~ mean(.x, na.rm = TRUE)))

context_wide <- context %>%
  select(o_net_soc_code, element_name, data_value) %>%
  pivot_wider(names_from = element_name, values_from = data_value) %>%
  group_by(o_net_soc_code) %>%
  summarise(across(everything(), ~ mean(.x, na.rm = TRUE)))
```

```r
ete_scaled <- scale(ete_wide[,-1])
context_scaled <- scale(context_wide[,-1])
```

```r
base_final <- knowledge_combined %>%
  left_join(skills_combined, by = "o_net_soc_code") %>%
  left_join(activities_combined, by = "o_net_soc_code") %>%
  left_join(ete_wide, by = "o_net_soc_code") %>%
  left_join(context_wide, by = "o_net_soc_code")
```

```r
base_numeric <- base_final %>%
  select(-o_net_soc_code)
```

```r
base_numeric <- base_numeric[, colMeans(is.na(base_numeric)) < 0.4]
```

```r
base_imputed <- base_numeric
base_imputed[,] <- apply(base_numeric, 2, function(x) {
  x[is.na(x)] <- median(x, na.rm = TRUE)
  return(x)
})
```

```r
base_scaled <- scale(base_imputed)
```
